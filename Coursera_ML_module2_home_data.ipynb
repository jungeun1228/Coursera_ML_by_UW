```python
import pandas as pd
```

#1) Load house sales data

```python
# Data reading (home_data.csv)
url = "https://drive.google.com/file/d/1f2zw6zjlaaxGyaSSN_4_CYdwXgVKOHY3/view?usp=sharing"
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
data = pd.read_csv(path)
```

#3) Simple regression model that predicts price from square feet

```python
from sklearn.model_selection import train_test_split
```

```python
#train/test split
X = sales_df.drop(columns = ['id','date'])
y = X.pop('price')  #pop: take out the column from original DF
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
#Feature: only sqft_living
X_sqft_train = pd.DataFrame(data = X_train['sqft_living'])
X_sqft_test = pd.DataFrame(data = X_test['sqft_living'])
```

train simple regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
sqft_model = LinearRegression()
```

```python
sqft_model.fit(X_sqft_train,y_train)
```

```python
sqft_predictions = sqft_model.predict(X_sqft_test)
```

#4) Evaluate the quality of our model

```python
from sklearn.metrics import root_mean_squared_error as rmse
sqft_rmse = rmse(y_true = y_test,
                 y_pred = sqft_predictions)
sqft_rmse
```

#5) Explore model a little further

```python
sqft_model.coef_
```

```python
sqft_model.intercept_
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(X_sqft_test,y_test,'.',
        X_sqft_test,sqft_model.predict(X_sqft_test),'-');
```

3. Building a regression model with several more features:  In the sample notebook, we built two regression models to predict house prices, one using just ‘sqft_living’ and the other one using a few more features, we called this set....

***답이 틀렸음, seed 가 달라서 인것으로 추정. 내꺼는 sklearn썼고 강사는 tripuricate 썼음, seed number도 다름, 등등 변수가 많음

```python
advanced_features = [
'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'zipcode',
'condition', # condition of house
'grade', # measure of quality of construction
'waterfront', # waterfront property
'view', # type of view
'sqft_above', # square feet above ground
'sqft_basement', # square feet in basement
'yr_built', # the year built
'yr_renovated', # the year renovated
'lat', 'long', # the lat-long of the parcel
'sqft_living15', # average sq.ft. of 15 nearest neighbors
'sqft_lot15', # average lot size of 15 nearest neighbors
]
```

```python
#Feature: multiple features
X_advanced_train = pd.DataFrame(data = X_train[advanced_features])
X_advanced_test = pd.DataFrame(data = X_test[advanced_features])
```

```python
#Build a new model with advanced features
advanced_model = LinearRegression()
advanced_model.fit(X_advanced_train,y_train)
advanced_predictions = advanced_model.predict(X_advanced_test)
```

```python
sqft_rmse = rmse(y_true = y_test,
                 y_pred = sqft_predictions)
multiple_rmse = rmse(y_true = y_test,
                 y_pred = multiple_predictions)
advanced_rmse = rmse(y_true = y_test,
                 y_pred = advanced_predictions)
print(sqft_rmse,multiple_rmse, advanced_rmse)
```
